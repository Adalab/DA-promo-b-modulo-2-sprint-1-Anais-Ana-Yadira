EJERCICIO 1: Cuántos ataques hubo por país? Devuelve un DataFrame.
Ordena los resultados de mayor a menor. Esto todavía no lo hemos visto, pero un DataFrame lo podemos ordenar con .sort_values()

```python
attack_pais = df.groupby(['country'])['case_number'].count()
attack_pais = pd.DataFrame(attack_pais)
attack_pais.reset_index(inplace = True)
attack_pais.sort_values(by='case_number')
```

EJERCICIO 2: ¿Cuántos ataques ha habido por año? Devuelve un DataFrame. De nuevo ordena los resultados de mayor a menor.

```python
attack_year = pd.DataFrame(df.groupby('year')['case_number'].count())
attack_year.reset_index(inplace=True)
attack_year.sort_values(by='case_number')
```

EJERCICIO 3: ¿Cuántos ataques hubo por año y sexo? Devolved un DataFrame solo del año 2000 en adelante.

```python
attack_year = pd.DataFrame(df.groupby(['year', 'sex_'])['case_number'].count())

attack_year.reset_index(inplace=True)
attack_y_sex = attack_year[attack_year['year']>= 2000].sort_values(by=['year','sex_'])
attack_y_sex
```

EJERCICIO 4: Primero extraed solo las filas que correspondan a USA, AUSTRALIA y SOUTH AFRICA. ¿Cuántos de los ataques fueron letales por sexo? Devuelve un DataFrame.

```python
paises = df[(df['country']=='USA') | (df['country']== 'AUSTRALIA') | (df['country']=='SOUTH AFRICA')]
paises
```

```python
paises_sexo = paises[paises['fatal_(y/n)']=='Y']
paises_sexo
```

```python
attack_paises = pd.DataFrame(paises_sexo.groupby(['sex_', 'country'])['fatal_(y/n)'].count())
attack_paises.reset_index(inplace=True)
attack_paises.sort_values(by=['country','fatal_(y/n)'])
```

EJERCICIO 5: Usando el DataFrame creado en el ejercicio anterior. ¿En qué país hubo más ataques?

```python
maximo = pd.DataFrame(attack_paises.groupby('country')['fatal_(y/n)'].max())
maximo.reset_index(inplace=True)
maximo
```
