EJERCICIO 1: Cread una columna nueva y una funciÃ³n que nos de el mes donde ocurriÃ³ el ataque. 
Tened en cuenta que no todas las filas tienen la misma estructura y que puede que no haya la informaciÃ³n de mes. 
En ese caso devolved un nulo (NaN).

```python
df_copia = df.copy()
```

```python
df_copia
```

```python
def sacar_mes(x):
    try:
        return re.findall('([A-Z]{1}[a-z]{2})\-',x)[0]
    except:

        return np.nan  
```

```python
df_copia["mes"] = df_copia["date"].apply(sacar_mes)
```

EJERCICIO 2: Vamos con la columna fatal_(y/n). Lo primero que tenÃ©is que hacer es evaluar los valores Ãºnicos que hay. EsperarÃ­amos tener solo dos valores: y y n, pero tristemente no deberÃ­a ser asÃ­ ðŸ˜­. Cread una columna nueva y una funciÃ³n que devuelva Ãºnicamente y o n y que devuelva un nulo (NaN) si no se encuentra el valor.

```python
df_copia['fatal_(y/n)'].unique()
```

```python

def fatal(fila):
    try:
       return re.findall(r'\b([N-zZ-a])\b', fila)[0]
    except:
        return np.nan


```

```python
df_copia["fatal_"] = df_copia["fatal_(y/n)"].apply(fatal)
df_copia
```

```python
df_copia['fatal_'].unique()
```

```python
df_copia['fatal_'].replace(to_replace = ['Y', 'N'], value = ['y', 'n'], inplace=True)
```

```python
df_copia['fatal_'].unique()
```

De nuevo en la columna hay otros carÃ¡cteres raros como 'M ', 'lli', 'N', '.'. En esos casos, queremos que se reemplacen por M o F de forma aleatoria (tendrÃ©is que usar la librerÃ­a random). En caso de que sea nulo, que la funciÃ³n devuelva nulo (np.nan).

```python
sex = df_copia['sex_'].unique()
type(sex[1])
```

```python
val_sexos = ['M', 'F']
```

```python

def sexo(x):

    try:
        np.isnan(x) == True
        return np.nan
   
    except:

        try:
            return re.findall(r'[F]|[f]|[M]|[m]', x)[0]
            
        except:

                return np.random.choice(val_sexos, size=1)[0]
```

```python
df_copia["sex2"] = df_copia["sex_"].apply(sexo)
```
